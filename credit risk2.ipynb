```python
# all continius varaibles must be numeric

rajat['emp_length'].unique()
rajat['emp_length'] = rajat['emp_length_int']
```

```python
# rajat['emp_length'] = rajat['emp_length_int']

rajat['emp_length_int'] = rajat['emp_length'].str.replace('+ years','')
rajat['emp_length_int'] = rajat['emp_length'].str.replace('+< 1 year',str(0))
rajat['emp_length_int'] = rajat['emp_length'].str.replace(' years',str(0))
```

```python
# convert str to integer
rajat['emp_length_int'] = pd.to_numeric(rajat['emp_length_int'], errors='coerce')
```

```python
rajat['earliest_cr_line_date'] = pd.to_datetime(rajat['earliest_cr_line'], format='%b-%y', errors='coerce')

# Fixing years that are incorrectly parsed (assuming dates before 2025 should be in the 1900s)
rajat.loc[rajat['earliest_cr_line_date'].dt.year > 2025, 'earliest_cr_line_date'] -= pd.DateOffset(years=100)
```
